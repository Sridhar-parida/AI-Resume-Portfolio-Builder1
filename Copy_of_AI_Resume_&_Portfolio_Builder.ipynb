```python

import pandas as pd
import torch
from transformers import pipeline
from sklearn.feature_extraction.text import TfidfVectorizer

```

```python
# Load a pre-trained text generation model
generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
# Student details (sample input)
student_data = {
    "name": "Rahul Kumar",
    "education": "B.Tech in Computer Science",
    "skills": "Python, Machine Learning, SQL, Data Analysis",
    "projects": "AI Resume Builder, Student Attendance System",
    "experience": "Fresher"
}
```

```python
# Job description
job_description = """
Looking for a Data Analyst with strong Python, Machine Learning,
and SQL skills. Experience with data visualization is a plus.
"""
```

```python
# Convert text to numerical features
vectorizer = TfidfVectorizer(stop_words='english')

# Combine student skills and job description
documents = [student_data["skills"], job_description]

# Fit and transform text
tfidf_matrix = vectorizer.fit_transform(documents)

# Get feature names
keywords = vectorizer.get_feature_names_out()
```

```python
# Create resume prompt
resume_prompt = f"""
Create a professional resume for the following student:

Name: {student_data['name']}
Education: {student_data['education']}
Skills: {student_data['skills']}
Projects: {student_data['projects']}
Experience: {student_data['experience']}

Make it ATS friendly.
"""
```

```python
# Generate resume
resume_output = generator(
    resume_prompt,
    max_length=250,
    num_return_sequences=1
)

# Print resume
print("AI GENERATED RESUME:\n")
print(resume_output[0]['generated_text'])
```

```python
# Create cover letter prompt
cover_letter_prompt = f"""
Write a professional cover letter for a student applying to a Data Analyst role.
Skills: {student_data['skills']}
Projects: {student_data['projects']}
"""
```

```python
# Generate cover letter
cover_letter_output = generator(
    cover_letter_prompt,
    max_length=200,
    num_return_sequences=1
)

# Print cover letter
print("\nAI GENERATED COVER LETTER:\n")
print(cover_letter_output[0]['generated_text'])
```

```python
# Save resume to text file
with open("AI_Resume.txt", "w") as file:
    file.write(resume_output[0]['generated_text'])

print("Resume saved successfully!")
```
